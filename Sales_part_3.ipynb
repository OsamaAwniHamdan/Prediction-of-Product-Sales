### Load dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fpath = "/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/Data/cleaned_sales_predictions_2023.csv"

df = pd.read_csv(fpath)
df = df.drop(columns=['Unnamed: 0'])
print(df.info())
df.head()
```

## Histograms to view the distributions of numerical features in your dataset.

#### Item Fat count

```python
ax = sns.histplot(x='Item_Fat_Content', data=df)
ax.set_title('Item Fat Count')
```

```python
sns.histplot(df['Item_Visibility'], kde=True)
```

#### Outlet location

```python
ax = sns.countplot(x='Outlet_Location_Type', data=df, hue='Item_Fat_Content')
ax.set_title('Outlet location type per item fat content')
```

```python
ax = sns.barplot(data=df, x='Outlet_Type', y='Item_Outlet_Sales', hue='Outlet_Location_Type')
ax.set_title('Outlet type vs Item_Outlet_Sales per outlet location')
plt.xticks(rotation=45, ha='right')
```

```python
sns.barplot(data=df, x='Outlet_Establishment_Year', y='Item_Outlet_Sales')
ax.set_title('Item outlet sales vs outlet establishment ,year')
```

```python
sns.barplot(data=df, x='Item_Type', y='Item_MRP', hue='Item_Fat_Content')
plt.xticks(rotation=45, ha='right')
```

#### Heatmap and correlation

```python

corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap = 'Greens', annot=True)
```

**Seems there is a moderate positive correlation  *Item_MRP* and *Item_Outlet_Sales***

#### drawing scatter plot between Item_MRP and Item_Outlet_Sales

```python
line_kws = dict(color='black')
scatter_kws = dict(edgecolor='white')

ax = sns.regplot(data=df, x='Item_MRP', y='Item_Outlet_Sales', line_kws=line_kws, scatter_kws=scatter_kws)

ax.set_title('Item_MRP vs Item_Outlet_Sales')
```

##### Check if item type affects the relation between Item MPR and Item outlet salse.

```python
line_kws = dict(color='black')
scatter_kws = dict(edgecolor='white')
g = sns.lmplot(data=df, x='Item_MRP', y='Item_Outlet_Sales', line_kws=line_kws, scatter_kws=scatter_kws, hue='Item_Type')
g.ax.set_title('Item_MRP vs Item_Outlet_Sales per item type')
```

The Item type does not affect the relation between outlet sals and MRP

#### Item_MRP vs Item_Weight
